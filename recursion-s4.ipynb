```python
# head Recursion 

# Time complexity = number of function calls =  O(N+1) = O(N)
# Space complexity = number of functions in stack space = O(N+1) = O(N)



count = 0

def greet():
    
    global count                           # always should be on top in the function 
    
    if count == 4:
        return
    
    print("Hello Arjun")
    
    
    count+=1
    greet()
    

greet()
```

```python
# Tale Recursion

count = 0

def greet():
    
    global count
    
    if count > 4 :
        return
    
    count +=1
    
    greet()
    
    print("Hello Arjun")
    
greet()
```

Recursion using parameters 

```python
# print x , N times 


def func(x,n):
    
    if n < 0:
        return
    
    print(x)
    func(x,n-1)


func(10,5)
```

```python
# print n to 1  { using tail or backtrack}


def func(i,n):
    
    if i > n:
        return 
    
    func(i+1,n)
    print(i)
    
func(1,10)



# using head Recursion 

# def func(n):
    
#     if n == 0:
#         return
    
#     print(n)
#     func(n-1)



# func(5)
```
