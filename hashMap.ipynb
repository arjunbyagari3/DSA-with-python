Store Frequency in Hash Map ( Dictionary )

```python
# method - 1 (recommended)

# Time complexity = O()
# Space complexity = O()


arr = [1,4,7,9,4,2,6,9,5,3,7,9,3]

n = len(arr)

freqMap = {}


for i in range(0,n):                                                # Time complexity = O(N)
    freqMap[arr[i]] = freqMap.get(arr[i],0)+1                       # Time complexity = O()             # Note: .get(key:any, default:any),  Return the value for key if key is in the dictionary, else default.
    
    
print(freqMap)


# method - 2 

# Time complexity = O(N)
# Space complexity = O(N)


# for i in range(0,n):
#     if arr[i] in freqMap:                                         # Time complexity = O(1)
#         freqMap[arr[i]] +=1                                       # Time complexity = O(1)
        
#     else:
#         freqMap[arr[i]] = 1

# print(freqMap)


print(freqMap[3])
```
