Store Frequency in Hash Map ( Dictionary )

```python
# method - 1 (recommended)

# Time complexity = O(N)
# Space complexity = O(N)


arr = [1,4,7,9,4,2,6,9,5,3,7,9,3]

n = len(arr)

freqMap = {}


for i in range(0,n):                                                # Time complexity = O(N)
    freqMap[arr[i]] = freqMap.get(arr[i],0)+1                       # Time complexity = O(1)             # Note: .get(key:any, default:any),  Return the value for key if key is in the dictionary, else default.
    
    
print(freqMap)


# method - 2 

# Time complexity = O(N)
# Space complexity = O(N)


# for i in range(0,n):
#     if arr[i] in freqMap:                                         # Time complexity = O(1)
#         freqMap[arr[i]] +=1                                       # Time complexity = O(1)
        
#     else:
#         freqMap[arr[i]] = 1

# print(freqMap)


print(freqMap[3])
```

Number hashing 

```python
# constrains 
# 1) 1 <= n[i] <= 10
# 2) arr1 can have 10**8 elements 
# 2) arr2 can have 10**8 elements 



# Brute force 

# Time complexity = O(N*N)                      # prone to TLE Error 
# Space complexity = O(1)


arr1 = [5,3,2,2,1,5,5,7,5,10]

arr2 = [10,111,1,9,5,67,2]



for i in range(0,len(arr2)):
    count = 0
    
    for j  in range(0,len(arr1)):
        if arr1[j] == arr2[i]:
            count+=1
            
    
    print(arr2[i],":",count)
```

```python
# Optimal 

# Time complexity = O(N+M)
# Space complexity = O(11) = O(1)



arr1 = [5,3,2,2,1,5,5,7,5,10]

arr2 = [10,111,1,9,5,67,2]


hashList = [0] * 11

for i in range(0,len(arr1)):
    hashList[arr1[i]] +=1
    

for j in range(0,len(arr2)):
    if arr2[j] < 1 or arr2[j] >10:
        print(0)
    else:
        print(arr2[j],":",hashList[arr2[j]])
    
hashList
```

```python
# constrains 
# 1) n[i] can be any number 
# 2) arr1 can have 10**8 elements 
# 2) arr2 can have 10**8 elements 


# using freqMap


arr1 = [5,3,2,2,1,5,5,7,5,10]

arr2 = [10,111,1,9,5,67,2]



freqMap = {}

for i in range(0,len(arr1)):
    
    freqMap[arr1[i]] = freqMap.get(arr1[i],0) + 1
    
    
for j in range(0,len(arr2)):
    if arr2[j] in freqMap:
        print(arr2[j],":",freqMap[arr2[j]])
    
    
```

Character Hashish 

```python
# constrains 
# 1) 'a' <= n[i] <= 'z'



# ASCII range --> 'a' to 'z' = 97 to 122

# Time complexity = O(N+M)
# Space complexity = O(26) =O(1)

s = "axyzxassyaaasxxx"

q = ["a","s","r","y"]

hashList = [0]*26


for i in range(0,len(s)):
    asc = ord(s[i])
    ind = asc - 97 
    hashList[ind]+=1
    
for j in range(0,len(q)):
    asc = ord(q[j])
    ind = asc - 97
    
    print(q[j],":",hashList[ind])
    
```
