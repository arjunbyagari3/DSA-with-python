```python
# brute force solution 

# Time complexity = O(N)
# Space complexity = O(k)               k = number of factors


num = 20 

facts = []

for i in range(1,num+1):
    if num % i == 0:                            # Time complexity of if condition is O(1)
        facts.append(i)                         # Time complexity append operation to list is O(1) because adding element at the end 
        
print(facts)
```

```python
# better solution 
# Time complexity = O(N/2) = O(N)
# space complexity = O(k)                 k = number of factors              

num = 30

facts = []


for i in range(1,(num//2)+1):
    if num % i == 0 :
        facts.append(i)
        
facts.append(num)

print(facts)
```

```python
# optimal solution 

# Time complexity = O(sqrt(N)) + O(N.log(N))
# space complexity = O(k)


num = 36 


facts = []

for i in range(1,int(num**0.5)+1):
    if num % i == 0 :
        facts.append(i)
    
    if num // i != i :
        facts.append(num//i)
        

# print(facts)

facts.sort()                     # Time complexity = O(N.log(N))                    # .sort() is an inplace method which modifies the original array(list) directly

print(facts)


















#! miscellaneous way to solve

# num = 36

# num = 30

# facts = []

# for i in range(1,int(num**0.5)+1):
#     a = num % i 
#     if a != 0 and a not in facts:
#         facts.append(a)
#     if i not in facts:
#         facts.append(i)
    
# print(facts)
```

2) Prime number check

```python
# num = 26

num = 11


def isPrime(num):
    for i in range(2,int(num**0.5)+1):
        if num % i == 0 :
            return False
        else:
            return True


isPrime(num)
```
