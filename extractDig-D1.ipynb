2) Count the Number of Digits in an Integer

```python
# 2) Count the Number of Digits in an Integer

# Time complexity = O(log(N) to base 10)                Note : dividing a the input with a number "a" will lead to the log N base a 
# Space complexity = O(1)


num = 123456

# num = 1

# num = 0

original = num

count = 0



while num > 0 :
    lastDigit = num % 10 
    count +=1
    num = num //10
    
    
if original == 0:
    print(1)
else:
    print(count)




# by using math library 

# import math


# print(int(math.log10(original)+1))
```

3) Reversing a number
4) Check if a Number is Palindrome or Not

```python
# Check if a Number is Palindrome or Not 

# Time complexity = O(log(N) base 10)
# space complexity = O(1)




num = 1234321

# num = 12

# num = 0


original = num

revNum = 0

while num > 0 :
    lastDigit = num % 10
    revNum = revNum*10 +lastDigit
    num = num //10
    
# print(revNum)

if original == revNum:
    print(True)
else:
    print(False)

```

5) Armstrong number check

```python
# Time complexity = O(log(N) base 10)
# space complexity = O(1)



# num = 153 

# num = 123

num = 1634

original = num

n = len(str(original))              # we can count the number of digits from the same pattern # 2nd program

sum = 0

while num > 0 :
    lastDigit = num % 10
    
    sum = sum + (lastDigit**n)
    
    num = num//10
    
    
# print(sum)

if sum == original:
    print(True)
    
else:
    print(False)
```
